# Example of Matplotlib
## Drawing the Brune's omega squared model
### - Color code
https://matplotlib.org/examples/color/named_colors.html

### <font color='red'>- Data preparation
</font> 

```python
###Drawing the omega squared model
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
import numpy as np
import matplotlib.cm as cm
import sys

Mo = 1.0
fc = 1.0
f = np.linspace(1,5,50)

A = 10**(f-3)

d = Mo/(1+(A/fc)**2)
v = Mo*A/(1+(A/fc)**2)
a = Mo*A*A/(1+(A/fc)**2)

#Normarize maximum Amplitude
d = d / Mo   #At longer period > seismic moment
a = a / (Mo*fc**2)      #At corner frequency
v = v / (Mo*fc/2.0)    #At Higher frequency
```

## <font color='red'>- Drawing the Figure
</font> 

```python
fig = plt.figure()
plt.plot(A, d, 'g-', A, v, 'r-', A, a, 'b-')
plt.xscale('log') 
plt.yscale('log')
plt.rcParams['font.size'] = 14
plt.xlabel('Frequency')
plt.ylabel('Foulier Spectrum')


#Preparation
#fig, ax1 = plt.subplots()
#For first data
#ax1.plot(A, v, 'g-')
#plt.yscale('log') 
#plt.ylabel('Foulier Velocity Spectrum')
#ax1.set_ylim(0.0001,1.5)
#For second data
#ax2=ax1.twinx()
#ax2.plot(A, a, 'b-')
#Adjustment
#plt.rcParams['font.size'] = 14
#plt.ylabel('Foulier Acceleration Spectrum')
#plt.xlabel('Frequency')
#plt.xscale('log') 
#plt.yscale('log') 
#ax2.set_ylim(0.0001,1.5)
#fig.savefig("out.ps")
plt.show()
```
